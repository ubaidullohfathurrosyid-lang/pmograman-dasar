```python
N = int(input("Masukkan angka N: "))

# Calculate the sum of even numbers
jumlah_genap = 0
for i in range(1, N + 1):
  if i % 2 == 0:  # Check for even numbersS
    jumlah_genap += i

print(f"Jumlah bilangan genap dari 1 sampai {N} adalah: {jumlah_genap}")

# Calculate the factorial
faktorial = 1
for i in range(1, N + 1):
  faktorial *= i

print(f"Faktorial dari {N} adalah: {faktorial}")
```
